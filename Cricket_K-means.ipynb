## Feature Scaling

Need to remove the players name values as it is object column and cant help in forming the clusters. We will remove the players name from cpy of the dataset

```python
df_drop=df.copy()
```

```python
df_drop
```

```python
df_drop.drop(['Player'],axis=1,inplace=True)
```

```python
#Standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scaled=sc.fit_transform(df_drop)
```

```python
df_scaled
```

```python
df_scaled=pd.DataFrame(df_scaled,columns=df_drop.columns)
```

```python
df_scaled
```

## Kmeans Model Implementation

```python
from sklearn.cluster import KMeans
```

```python
#Checking elbow plot to get optimal no. of clusters
K_values = list(range(2,8))
```

```python
K_values
```

```python
#max_iter is basically the iterations till the number it will stop iterating the centroid for the clusters
ssd=[]
for K in K_values:
    Kmodel=KMeans(n_clusters=K, max_iter=150, random_state=23)
    Kmodel.fit(df_scaled)
    ssd.append(Kmodel.inertia_)
```

```python
K_values
```

```python
ssd
```

```python
plt.plot(K_values, ssd)
plt.show()
```

```python
#Final model with 4 clusters will be most optimal
final_model = Kmodel=KMeans(n_clusters=4, max_iter=150, random_state=23)
final_model.fit(df_scaled)
```

```python
final_model.labels_
```

```python
df['Cluster ID']= final_model.labels_
```
